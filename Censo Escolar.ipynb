```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
censo_df = pd.read_csv('dados/censo_escolar/censoevolutivotodasasredes.csv', index_col='CODEOL', encoding='latin-1', sep=';')
```

```python
censo_df
```

```python
censo_df.groupby(['ANO', 'DEP'])[['AL_CRECHE', 'AL_PRE']].sum()
```

```python
censo_df.loc[censo_df['DEP'].isin(['MUNICIPAL', 'CONVMUN', 'Municipal'])].groupby(['ANO', 'DEP'])[['AL_CRECHE', 'AL_PRE']].sum()
```

```python
evol_ens_inf_df = censo_df.loc[censo_df['DEP'].isin(['MUNICIPAL', 'CONVMUN', 'Municipal'])].groupby(['ANO'])[['AL_CRECHE', 'AL_PRE']].sum()
```
